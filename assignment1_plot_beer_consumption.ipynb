```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('istherecorrelation.csv', sep=";")
data.columns = data.columns.str.strip()
data["WO [x1000]"] = (
    data["WO [x1000]"]
    .astype(str)
    .str.replace(",", ".", regex=False)
    .astype(float)
)
print(data)
```

```python
year = data["Year"]
num_students = data["WO [x1000]"]
beer = data["NL Beer consumption [x1000 hectoliter]"]

beer_per_student = beer / num_students
print(beer_per_student)
```

```python
plt.figure(figsize=(10, 6), dpi=300)
plt.plot(year, beer_per_student, marker="o", linestyle="-", color="hotpink")

plt.title("Average Yearly Beer Consumption per University Student in the Netherlands")
plt.xlabel("Year")
plt.ylabel("Beer per Student (liters)")
plt.grid(True, linestyle=":")
plt.show()
```
